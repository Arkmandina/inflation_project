```python
import pandas as pd

file_data = pd.read_csv("./Data/Global Dataset of Inflation.csv", encoding="latin1")
file_data.head()
```

```python
file_data.dtypes
```

```python
clean_data = file_data[["Country Code", "Country", "2005","2006", "2007", "2008", "2009", "2010","2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2021", "2022"]]
clean_data
```

```python
lower_clean_data = clean_data.rename(columns = {"Country Code":"country code", "Country":"country"})
lower_clean_data
```

```python
final_data = lower_clean_data.dropna()
final_data
```

```python
final_data.to_csv('Clean_Global Dataset of Inflation.csv')
```
